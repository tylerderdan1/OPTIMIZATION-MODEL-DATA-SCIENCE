```python
import matplotlib.pyplot as plt
from pulp import *

# --- Problem 1: Product Mix Optimization ---
model1 = LpProblem("Maximize_Profit", LpMaximize)

x = LpVariable("Product_A", lowBound=0, cat='Continuous')
y = LpVariable("Product_B", lowBound=0, cat='Continuous')

model1 += 20 * x + 30 * y, "Total_Profit"
model1 += 2 * x + 1 * y <= 100, "Machine_1_Time"
model1 += 1 * x + 2 * y <= 80, "Machine_2_Time"

model1.solve()

print("üìå Product Mix Optimization Results:")
print(f"Status: {LpStatus[model1.status]}")
print(f"Produce Product A: {x.varValue}")
print(f"Produce Product B: {y.varValue}")
print(f"Maximum Profit: ‚Çπ{value(model1.objective)}\n")

# Visualization: Product Mix
products = ['Product A', 'Product B']
quantities = [x.varValue, y.varValue]

plt.figure(figsize=(6,4))
plt.bar(products, quantities, color=['skyblue', 'orange'])
plt.title('Optimal Production Quantities')
plt.ylabel('Units to Produce')
plt.show()


# --- Problem 2: Transportation Optimization ---
warehouses = ['A', 'B']
stores = ['1', '2', '3']

supply = {'A': 70, 'B': 50}
demand = {'1': 40, '2': 50, '3': 30}

costs = {
    ('A', '1'): 2, ('A', '2'): 4, ('A', '3'): 5,
    ('B', '1'): 3, ('B', '2'): 1, ('B', '3'): 7
}

model2 = LpProblem("Minimize_Transportation_Cost", LpMinimize)

x = LpVariable.dicts("route", (warehouses, stores), lowBound=0, cat='Continuous')

model2 += lpSum(x[w][s] * costs[(w, s)] for w in warehouses for s in stores), "Total_Cost"

for w in warehouses:
    model2 += lpSum(x[w][s] for s in stores) <= supply[w], f"Supply_{w}"
for s in stores:
    model2 += lpSum(x[w][s] for w in warehouses) == demand[s], f"Demand_{s}"

model2.solve()

print("üìå Transportation Optimization Results:")
print(f"Status: {LpStatus[model2.status]}")

shipments = []
for w in warehouses:
    for s in stores:
        qty = x[w][s].varValue
        shipments.append((w, s, qty))
        print(f"Ship from Warehouse {w} to Store {s}: {qty} units")

print(f"Total Minimum Cost: ‚Çπ{value(model2.objective)}")

# Visualization: Shipments by Warehouse
import numpy as np

fig, ax = plt.subplots(figsize=(8,5))
bar_width = 0.35
index = np.arange(len(stores))


warehouse_A_shipments = [qty for w,s,qty in shipments if w == 'A']
warehouse_B_shipments = [qty for w,s,qty in shipments if w == 'B']

bar1 = ax.bar(index, warehouse_A_shipments, bar_width, label='Warehouse A', color='teal')
bar2 = ax.bar(index + bar_width, warehouse_B_shipments, bar_width, label='Warehouse B', color='coral')

ax.set_xlabel('Stores')
ax.set_ylabel('Units Shipped')
ax.set_title('Units Shipped from Warehouses to Stores')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(stores)
ax.legend()

plt.show()
```
